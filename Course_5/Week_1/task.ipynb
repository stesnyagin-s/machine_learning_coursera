```python
import pandas as pd
import statsmodels.api as sm
import numpy as np
```

```python
milk = pd.read_csv('monthly-milk-production.csv',';', index_col=['month'], parse_dates=['month'], dayfirst=True)
milk.head()
```

```python
milk.milk.plot()
```

```python
sm.tsa.stattools.adfuller(milk.milk)
```

```python
milk["daily"] = milk.milk/(milk.index.days_in_month)
```

```python
milk.daily.plot()
```

```python
np.sum(milk.daily)
```

```python
milk["daily_diff1"] = milk.daily - milk.daily.shift(1)
milk["daily_diff12"] = milk.daily - milk.daily.shift(12)
milk["daily_diff122"] = milk.daily_diff12 - milk.daily_diff12.shift(1)
```

```python
sm.tsa.stattools.adfuller(milk["daily_diff1"][1:])
```

```python
sm.tsa.stattools.adfuller(milk["daily_diff12"][12:])
```

```python
sm.tsa.stattools.adfuller(milk["daily_diff123"][24:])
```

```python
sm.tsa.stattools.adfuller(milk["daily_diff122"][13:])
```

```python
import statsmodels.api as sm
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(20, 1))
sm.graphics.tsa.plot_acf(milk["daily_diff122"][13:].values.squeeze(), lags=50)
plt.show()
```
