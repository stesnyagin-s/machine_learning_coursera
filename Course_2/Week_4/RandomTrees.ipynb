```python
from sklearn.datasets import load_digits
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = load_digits()
```

```python
X = dataset.data
y = data.target
```

```python
clf = DecisionTreeClassifier()
```

```python
answer1 = cross_val_score(clf, X, y, cv=10).mean()
```

```python
with open("1.txt", "w") as fout:
    fout.write(str(answer1))
```

```python
from sklearn.ensemble import BaggingClassifier
bagging_clf = BaggingClassifier(clf, n_estimators=100)
answer2 = cross_val_score(bagging_clf, X, y, cv=10).mean()
answer2 
```

```python
with open("2.txt", "w") as fout:
    fout.write(str(answer2))
```

```python
bagging_clf = BaggingClassifier(clf, n_estimators=100, max_features=int(np.sqrt(X.shape[1])))
answer3 = cross_val_score(bagging_clf, X, y, cv=10).mean()
answer3
```

```python
with open("3.txt", "w") as fout:
    fout.write(str(answer3))
```

```python
bagging_clf = BaggingClassifier(DecisionTreeClassifier(max_features='sqrt'), n_estimators=100)
answer4 = cross_val_score(bagging_clf, X, y, cv=10).mean()
answer4
```

```python
with open("4.txt", "w") as fout:
    fout.write(str(answer4))
```

```python
from sklearn.ensemble import RandomForestClassifier
m = []
n_trees = range(5,100,5)
for i in n_trees:
    m.append(cross_val_score(RandomForestClassifier(n_estimators = i), X, y, cv=10).mean())
plt.plot(n_trees, m)
```

```python
m = []
n_features = range(1,64)
for i in n_features:
    m.append(cross_val_score(RandomForestClassifier(max_features = i), X, y, cv=5).mean())
plt.plot(n_features, m)
```

```python
m = []
n_depth = range(1,30)
for i in n_depth:
    m.append(cross_val_score(RandomForestClassifier(max_depth = i), X, y, cv=10).mean())
plt.plot(n_depth, m)
```
