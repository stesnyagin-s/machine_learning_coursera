```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

# Экспоненциальное распределение
https://ru.wikipedia.org/wiki/Экспоненциальное_распределение

$ X \sim exp(\lambda) $<br>
$\lambda = \frac{1}{2}$<br>
$f(x) = \lambda \cdot exp(-\lambda \cdot x) = \frac{1}{2}\cdot exp(-\frac{1}{2}\cdot x) $ <br>
$M[X] = \lambda^{-1} = \frac{1}{2}^{-1} = 2 $ <br>
$D[X] = \lambda^{-2} = \frac{1}{2}^{-2} = 4 $

Создаём случайную величину имеющую экспоненциальное распределение. $ scale = \lambda^{-1} $

```python
exp_rv = sts.expon(loc=0, scale=2)
```

Генерируем выборку, строим гистограмму и теоретическую плотность распределения.

```python
sample = exp_rv.rvs(size=1000)
plt.hist(sample, label='Гистограмма выборки', density = True, bins = 30)
x = np.linspace(0,10,200)
pdf = exp_rv.pdf(x)
plt.plot(x, pdf,'r', label='Теоретическая плотность')
plt.ylabel('Частота \\ f(x)')
plt.xlabel('$x$')
plt.legend()
```

Для следующих трех случаев $( n = 5,10,50 )$, создаётся выборка средних выборочных для трех разных $n$. Строится гистограмма данной выборки и теоретическая плотность выборочного среднего.

## n = 5
$ M[\overline{X}] = 2$<br>
$ D[\overline{X}] = \frac{4}{n} = \frac{4}{5} $

```python
means = []
n = 5
for i in range(1000):
    sample = exp_rv.rvs(size=n)
    mean = sum(sample)/n
    means.append(mean)
mean = 2
var = 4/n
norm_rv = sts.norm(mean, np.sqrt(var))
x = np.linspace(0,5,200)
pdf = norm_rv.pdf(x)
plt.plot(x, pdf,'r', label='Теоретическая плотность')
plt.hist(means, density = True, bins = 20, label= 'Гистограмма выборки')
plt.ylabel('Частота \\ f(x)')
plt.xlabel('$x$')
plt.legend(loc='upper right')
```

## n = 10
$ M[\overline{X}] = 2 $ <br>
$ D[\overline{X}] = \frac{4}{n} = \frac{4}{10} $

```python
means = []
n = 10
for i in range(1000):
    sample = exp_rv.rvs(size=n)
    mean = sum(sample)/n
    means.append(mean)
norm_rv = sts.norm(2, np.sqrt(4/n))
x = np.linspace(0,10,200)
pdf = norm_rv.pdf(x)
plt.plot(x, pdf,'r', label='Теоретическая плотность')
plt.hist(means, density = True, bins = 20, label= 'Гистограмма выборки')
plt.ylabel('Частота \\ f(x)')
plt.xlabel('$x$')
plt.legend(loc='upper right')
```

## n = 50
$ M[\overline{X}] = 2 $ <br>
$ D[\overline{X}] = \frac{4}{n} = \frac{4}{50} $

```python
means = []
n = 50
for i in range(1000):
    sample = exp_rv.rvs(size=n)
    mean = sum(sample)/n
    means.append(mean)
mean = 2.
var = 4./n
norm_rv = sts.norm(mean, np.sqrt(var))
x = np.linspace(0,7,200)
pdf = norm_rv.pdf(x)
plt.plot(x, pdf,'r', label='Теоретическая плотность')
plt.hist(means, density = True, bins = 20, label= 'Гистограмма выборки')
plt.ylabel('Частота \\ f(x)')
plt.xlabel('$x$')
plt.legend(loc='upper right')
```

# Вывод
С увеличением $n$ дисперсия распределения выборочного среднего уменьшается и оно дает наиболее точную оценку истинного значения среднего(мат ожидания) случайной величины.
